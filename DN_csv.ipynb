## Построение ДН из csv файлов

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd

import scipy
```

```python
file_name = 'DataFold/DN.csv'
data = pd.read_csv(file_name, header=None, skiprows=2, delimiter='\s+', 
                   names=['Theta',  'Phi',  'Abs', 'Abs(Theta)', 'Phase(Theta)', 'Abs(Phi)', 'Phase(Phi)',  '1/Ax.Ratio'])

data
```

```python
FF = data[['Abs', 'Theta', 'Phi']]

FF[FF['Phi']==0]
```

```python
# fig, ax = plt.subplots(figsize = (12,6))

FF[FF['Phi']==0].plot(x='Theta', y='Abs', figsize=(12,6))

ax = plt.gca()

ax.set_ylim(ymin=-20, ymax=0)
ax.set_xticks(np.arange(-180, 181, 20))
ax.set_yticks(np.arange(-20, 10, 2.5))
ax.grid(linestyle='--', color='green')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('top')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```
